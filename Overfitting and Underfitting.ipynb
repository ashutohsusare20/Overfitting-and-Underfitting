# Overfitting and Underfitting

```python
Suppose we want to check how well our machine learning model learns and generalizes to the new data.

For that, we have overfitting and underfitting
```

## Bias :-

```python
Bias is the error rate of training data (simply error ==> Yp-Ya)

When error rate is high value then ----> is called as High Bias (high accuracy) 
when error rate is low value then ---->  is called as low Bias (low accuracy)
```

## Variance :- 

```python
It is the difference between the error rate of training data and error rate of testing data.

If diff is high --> High Variance
If diff is low ---> low variacne
```

#### for generalized or best model need both Bais and variance should be low

## 1) Overfitting

```python
Model is performinig well on only training datasets ( for the both classifi, regression)

Training Data Accuracy = 98%(classifiaction), 0.95( r2 - regression)
Training Data Accuracy = 70%(classifiaction), 0.69( r2 - regression) 

>> If datasets having low bais and high variance then >>>> Overfitting
```

```python
Reasons for Overfitting are as follows:
    
1) High variance and low bias 
2) The model is too complex
3) The size of the training data
```

### How to Avoid Overfitting

```python
#1st 5 is  impo
1. Increse Training Data
    
2. Early Stopping during the training phase

3. Reduce Number of Features(30 >> 14 , features/parameters ):
    Feature Selection Technique
    Dimensionality Reduction Techniques (PCA,LDA)
    
4. Reduce model complexity.

5. Remove Outliers
    
6. Cross Validation:
    k-fold Cross Validation
    
7. Regularization(L1 and L2): Only used in linear model(Linear Regresion and lOgistic Regression)
    1. L1 Regularization >> Lasso Regression
    2. L2 Regularization >> Ridge Regression

8. Hyperparameter Tuning:
    GridSearchCV
    RandomizedSearchCV

9. Pruning (Cutting Branches of Decision Tree)>> Tree Based Model
```

## Underfitting

```python
Model is neither performing well on training dataset nor on testing dataset

Training Data Accuracy = 70%(classification) , 0.55(r2 - Regression)
Testing Data Accuracy = 68%(classification) , 0.56(r2 - Regression)

>>> High Bias and Low Variance
```

```python
Reasons for Underfitting:

High bias and low variance 
The size of the training dataset used is not enough.
The model is too simple.
Training data is not cleaned and also contains noise in it.
```

## How to Avoid Underfitting

```python
1. Increase the number of features
    10 >> 15
    1. Add new features
    2. Add Derived Features(Creating multiple features from single feature)
    
2. Use Proper Feature Selection:
    1. Filter Methods
    2. Wrapper Method
    3. Embedded Method
    
3. Increase model complexity

4. Remove noise from the data

5. Hyperparameter Tuning
```
